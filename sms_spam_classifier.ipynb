## Step 3 . Data Preprocessing or Text PreProcessing 

```python
### Stpes involve in text pre processing 
##1. Lower Case 
##2. Tokenization 
##3. Removing special character 
##4. Removing Punctuation or Stop Words 
##5. Stemming 
```

```python
### Importing Required Libraries for text pre processing 

import string
from nltk.corpus import stopwords 
from nltk.stem import WordNetLemmatizer
```

```python
### Dealing with stop words and Applying Lemmatization 
 
stopwords = nltk.corpus.stopwords.words("english")
lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
### Text Preprocessing 

def text_preprocessing(text):
    text = text.lower()
    tokens = nltk.word_tokenize(text)
    tokens = [token for token in tokens if token.isalnum()]
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    clean_text = " ".join(tokens)
    return clean_text
```

```python
df["processed_text"] = df["Text"].apply(text_preprocessing)
```

## Step 4. Model Building 

```python
### Using Navie Bayes Importing required libraries 

from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
cv = CountVectorizer()
tfidf = TfidfVectorizer(max_features=4000)
```

```python
X = tfidf.fit_transform(df["processed_text"]).toarray()
```

```python
X
```

```python
X.shape
```

```python
y = df["Target"].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Checking the shape of the training and test sets

print('Shape of the training input data:', X_train.shape)
print('Shape of the training output data:', y_train.shape)
print('Shape of the test input data:', X_test.shape)
print('Shape of the test output data:', y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gnb = GaussianNB()
mub =  MultinomialNB()
bub = BernoulliNB()
```

```python
### Using Gaussian Naive bayes

gnb.fit(X_train, y_train)
y_pred1 = gnb.predict(X_test)
```

```python
accuracy1 = accuracy_score(y_test,y_pred1)
accuracy1
```

```python
confusion1 = confusion_matrix(y_test,y_pred1)
confusion1
```

```python
precision1 = precision_score(y_test,y_pred1)
precision1
```

```python
### Using Multinomial Navie Bayes 

mub.fit(X_train, y_train)
y_pred2 = mub.predict(X_test)
accuracy2 = accuracy_score(y_test,y_pred2)
print(accuracy2)
confusion2 = confusion_matrix(y_test,y_pred2)
print(confusion2)
precision2 = precision_score(y_test,y_pred2)
print(precision2)
```

```python
### Using Burnoli  Navie Bayes 

bub.fit(X_train, y_train)
y_pred3 = bub.predict(X_test)
accuracy3 = accuracy_score(y_test,y_pred3)
print(accuracy3)
confusion3 = confusion_matrix(y_test,y_pred3)
print(confusion2)
precision3 = precision_score(y_test,y_pred3)
print(precision3)
```

```python
# Define the board
board = [" " for i in range(9)]

# Print the board
def print_board():
    print("   |   |   ")
    print(" " + board[0] + " | " + board[1] + " | " + board[2] + " ")
    print("   |   |   ")
    print("-----------")
    print("   |   |   ")
    print(" " + board[3] + " | " + board[4] + " | " + board[5] + " ")
    print("   |   |   ")
    print("-----------")
    print("   |   |   ")
    print(" " + board[6] + " | " + board[7] + " | " + board[8] + " ")
    print("   |   |   ")

# Check if a player has won
def check_win(player):
    if (board[0] == player and board[1] == player and board[2] == player) or \
        (board[3] == player and board[4] == player and board[5] == player) or \
        (board[6] == player and board[7] == player and board[8] == player) or \
        (board[0] == player and board[3] == player and board[6] == player) or \
        (board[1] == player and board[4] == player and board[7] == player) or \
        (board[2] == player and board[5] == player and board[8] == player) or \
        (board[0] == player and board[4] == player and board[8] == player) or \
        (board[2] == player and board[4] == player and board[6] == player):
        return True
    else:
        return False

# Play the game
def play_game():
    print("Welcome to Tic Tac Toe!")
    print_board()
    player = "X"
    while True:
        print("Player " + player + ", it's your turn.")
        move = int(input("Enter a number between 1 and 9: ")) - 1
        if board[move] == " ":
            board[move] = player
        else:
            print("That space is already taken. Try again.")
            continue
        print_board()
        if check_win(player):
            print("Congratulations, player " + player + " has won!")
            break
        if " " not in board:
            print("It's a tie!")
            break
        player = "O" if player == "X" else "X"

# Start the game
play_game()
```
